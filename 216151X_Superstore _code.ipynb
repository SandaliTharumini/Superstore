# Data Preprocessing

Depending on kid/teen home, we will change Alone, Absurd, YOLO or others to either Single or Married

```python
df['Marital_Status'] = df.apply(lambda x: 'Married'
                                if ((x['Marital_Status'] not in ['Single', 'Married', 'Divorced', 'Together', 'Widow'])
                                    & ((x['Kidhome'] > 0) | (x['Teenhome'] > 0)))
                                else x['Marital_Status'],
                                axis=1)

df['Marital_Status'] = df.apply(lambda x: 'Single'
                                if ((x['Marital_Status'] not in ['Single', 'Married', 'Divorced', 'Together', 'Widow'])
                                    & ((x['Kidhome'] == 0) & (x['Teenhome'] == 0)))
                                else x['Marital_Status'],
                                axis=1)

df['Marital_Status'].value_counts()
```

In education, 2n cycle and Master means the same thing. We can combine these two categories.

```python
# Replacing 2n Cycle with Master
df["Education"] = df["Education"].replace("2n Cycle", "Master")
df['Education'].value_counts()
```

### Missing Value Imputation

```python

import scipy.stats as stats

# Plot histogram and KDE for Income with blue touch
plt.figure(figsize=(8, 6))
sns.histplot(df['Income'], kde=True, bins=30, color='#000080')
plt.title('Income Distribution with KDE')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()

# QQ plot for Income with blue touch
plt.figure(figsize=(6, 6))
stats.probplot(df['Income'].dropna(), dist="norm", plot=plt)
plt.title('QQ Plot for Income')

# Customize the QQ plot color
plt.gca().get_lines()[0].set_color('blue')
plt.gca().get_lines()[1].set_color('blue')

plt.show()
```

```python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test for Income
stat, p_value = shapiro(df['Income'].dropna())
print(f'Shapiro-Wilk Test: Statistic={stat}, p-value={p_value}')

if p_value > 0.05:
    print('Income is likely normally distributed (fail to reject H0)')
else:
    print('Income is not normally distributed (reject H0)')
```

```python
from scipy.stats import kstest

# Perform Kolmogorov-Smirnov test for Income
stat, p_value = kstest(df['Income'].dropna(), 'norm', args=(df['Income'].mean(), df['Income'].std()))
print(f'Kolmogorov-Smirnov Test: Statistic={stat}, p-value={p_value}')

if p_value > 0.05:
    print('Income is likely normally distributed (fail to reject H0)')
else:
    print('Income is not normally distributed (reject H0)')
```

since the income does not follows normal distribution median used to impute missing values.

We can update the missing Income with the median income of the education level.

```python
# median income by education level
incomebyEducation = df.groupby('Education')['Income'].median().to_dict()

# update missing income
df['Income']=df['Income'].fillna(df['Education'].map(incomebyEducation))
```

There are columns that have some outliers.

```python
# List of columns with identified outliers
columns_with_outliers = ['MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases','NumWebVisitsMonth','Income','Age']

def handle_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Clip the outliers to the lower and upper bounds
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

# Apply IQR method only to columns with outliers
for col in columns_with_outliers:
    handle_outliers_iqr(df, col)
```

The outliers in each numerical column were handled by clipping their values to the lower and upper bounds calculated using the IQR method, replacing extreme values with the nearest valid thresholds.

```python
def plot_pie_chart(data, column, title):
    # Mapping encoded values to original labels
    education_mapping = {0: 'Basic', 1: 'Graduation', 2: 'Master', 3: 'PhD'}

    # Apply mapping directly for the pie chart labels
    counts = data[column].value_counts()
    labels = [education_mapping[val] for val in counts.index]  # Get original labels for the pie chart

    # Use Seaborn's built-in color-blind-friendly palette
    colors = sns.color_palette("colorblind", n_colors=len(labels))

    plt.figure(figsize=(8, 5))

    # Create pie chart
    wedges, texts, autotexts = plt.pie(
        counts,
        labels=labels,
        autopct=lambda p: f'{p:.1f}%\n({int(p * sum(counts) / 100)})',
        startangle=90,
        colors=colors,
    )

    # Set text properties for autotexts
    for text in autotexts:
        text.set_color('white')  # Set the color of the percentage text to white

    plt.title(title, fontsize=13.5)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    # Add legend integrated into the pie chart
    plt.legend(
        labels,
        loc='center left',  # Adjust the location as needed
        title="Education Levels",  # Title for the legend
        frameon=False,  # No frame around the legend
        bbox_to_anchor=(1, 0, 0.5, 1)  # Adjusts the position of the legend
    )

    plt.show()

# Pie chart for Education with actual labels in the legend
plot_pie_chart(df, 'Education', 'Distribution of Education Levels')
```

# Handle Class Inbalance

As we have many values ​​of a class and few values ​​of others, we will balance the classes using OverSampling to make the classes balanced.

```python
from imblearn.over_sampling import RandomOverSampler

# Define features and target variable
X = df.drop(columns=['Response'])  # Features
y = df['Response']  # Target variable

# Use RandomOverSampler to balance the classes
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Check the new class distribution
print("New class distribution:")
print(y_resampled.value_counts())
```

```python
X_resampled.shape, y_resampled.shape
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets (70-30 split)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Check the shapes of the splits
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the training set and transform both training and testing sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Model Building

## Feature Selection

```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold

# Initialize the Random Forest model
rf = RandomForestClassifier(random_state=42)

# Initialize RFECV with cross-validation
rfecv_rf = RFECV(estimator=rf, step=1, cv=StratifiedKFold(5), scoring='accuracy')

# Fit RFECV on the training data
rfecv_rf.fit(X_train_scaled, y_train)

# Get the optimal number of features and selected features
print("Optimal number of features for Random Forest:", rfecv_rf.n_features_)
selected_features_rf = X_train.columns[rfecv_rf.support_]
print("Selected features for Random Forest:", selected_features_rf)
```

```python
# Get the feature importances from the fitted Random Forest model
importances_rf = rfecv_rf.estimator_.feature_importances_

# Sort the importances and corresponding feature names
indices = np.argsort(importances_rf)[::-1]
selected_features_rf_sorted = selected_features_rf[indices]

# Create a seaborn colorblind palette
color_palette = sns.color_palette("colorblind", len(selected_features_rf))

# Plot the feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x=importances_rf[indices], y=selected_features_rf_sorted, palette=color_palette)
plt.title('Feature Importances for Random Forest (Selected Features)')
plt.xlabel('Importance')
plt.ylabel('Feature')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
from xgboost import XGBClassifier

# Initialize the XGBoost model
xgb = XGBClassifier(random_state=42)

# Initialize RFECV with cross-validation
rfecv_xgb = RFECV(estimator=xgb, step=1, cv=StratifiedKFold(5), scoring='accuracy')

# Fit RFECV on the training data
rfecv_xgb.fit(X_train_scaled, y_train)

# Get the optimal number of features and selected features
print("Optimal number of features for XGBoost:", rfecv_xgb.n_features_)
selected_features_xgb = X_train.columns[rfecv_xgb.support_]
print("Selected features for XGBoost:", selected_features_xgb)
```

```python
# Get the feature importances from the fitted XGBoost model
importances_xgb = rfecv_xgb.estimator_.feature_importances_

# Sort the importances and corresponding feature names
indices = np.argsort(importances_xgb)[::-1]
selected_features_xgb_sorted = selected_features_xgb[indices]

# Create a seaborn colorblind palette
color_palette = sns.color_palette("colorblind", len(selected_features_xgb_sorted))

# Plot the feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x=importances_xgb[indices], y=selected_features_xgb_sorted, palette=color_palette)
plt.title('Feature Importances for XGBoost (Selected Features)')
plt.xlabel('Importance')
plt.ylabel('Feature')

# Show the plot
plt.tight_layout()
plt.show()
```

## Model Training

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

# Common selected features
common_features = ['Education', 'Income', 'Kidhome','Teenhome', 'Recency', 'MntWines', 'MntMeatProducts', 'MntGoldProds',
                   'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
                   'NumWebVisitsMonth', 'Participation_time', 'Marital_Status_Single',
                   'Marital_Status_Together', 'Marital_Status_Widow']

# ---- Decision Tree ----
# Use the common set of features for Decision Tree
X_train_common = X_train_scaled[:, [X_train.columns.get_loc(f) for f in common_features]]
X_test_common = X_test_scaled[:, [X_test.columns.get_loc(f) for f in common_features]]

# Train and evaluate the Decision Tree model
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_common, y_train)

# Predict and evaluate
dt_train_predictions = dt.predict(X_train_common)
dt_test_predictions = dt.predict(X_test_common)

# Accuracy for Decision Tree
dt_train_accuracy = accuracy_score(y_train, dt_train_predictions)
dt_test_accuracy = accuracy_score(y_test, dt_test_predictions)
print(f"Decision Tree Training Accuracy: {dt_train_accuracy:.4f}")
print(f"Decision Tree Testing Accuracy: {dt_test_accuracy:.4f}")
print("Decision Tree Classification Report:")
print(classification_report(y_test, dt_test_predictions))

# ---- Random Forest ----
# Use the common set of features for Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_common, y_train)

# Predict and evaluate
rf_train_predictions = rf.predict(X_train_common)
rf_test_predictions = rf.predict(X_test_common)

# Accuracy for Random Forest
rf_train_accuracy = accuracy_score(y_train, rf_train_predictions)
rf_test_accuracy = accuracy_score(y_test, rf_test_predictions)
print(f"Random Forest Training Accuracy: {rf_train_accuracy:.4f}")
print(f"Random Forest Testing Accuracy: {rf_test_accuracy:.4f}")
print("Random Forest Classification Report:")
print(classification_report(y_test, rf_test_predictions))

# ---- XGBoost ----
# Use the common set of features for XGBoost
xgb = XGBClassifier(random_state=42)
xgb.fit(X_train_common, y_train)

# Predict and evaluate
xgb_train_predictions = xgb.predict(X_train_common)
xgb_test_predictions = xgb.predict(X_test_common)

# Accuracy for XGBoost
xgb_train_accuracy = accuracy_score(y_train, xgb_train_predictions)
xgb_test_accuracy = accuracy_score(y_test, xgb_test_predictions)
print(f"XGBoost Training Accuracy: {xgb_train_accuracy:.4f}")
print(f"XGBoost Testing Accuracy: {xgb_test_accuracy:.4f}")
print("XGBoost Classification Report:")
print(classification_report(y_test, xgb_test_predictions))
```

```python
# ---- Plot feature importances ----
def plot_feature_importance(model, features, model_name):
    importances = model.feature_importances_
    indices = np.argsort(importances)[::-1]

    # Set the color-blind-friendly palette
    sns.set_palette("colorblind")

    plt.figure(figsize=(10, 6))
    plt.title(f"{model_name} Feature Importance", fontsize=16)
    plt.barh(range(len(indices)), importances[indices], align="center", color=sns.color_palette("colorblind"))
    plt.yticks(range(len(indices)), [features[i] for i in indices])
    plt.xlabel("Feature Importance Score")
    plt.ylabel("Features")
    plt.tight_layout()
    plt.gca().invert_yaxis()  # Invert y-axis to have the most important feature at the top
    plt.show()

# Plot for Decision Tree using common features
plot_feature_importance(dt, common_features, "Decision Tree")

# Plot for Random Forest using common features
plot_feature_importance(rf, common_features, "Random Forest")

# Plot for XGBoost using common features
plot_feature_importance(xgb, common_features, "XGBoost")
```

## Performance Evaluation

```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, balanced_accuracy_score,
                             matthews_corrcoef, cohen_kappa_score, log_loss, confusion_matrix, roc_curve, auc)
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score

# ---- Predictions for each model using the common feature set ----
# Decision Tree
dt_pred = dt.predict(X_test_scaled[:, [X.columns.get_loc(f) for f in common_features]])
dt_proba = dt.predict_proba(X_test_scaled[:, [X.columns.get_loc(f) for f in common_features]])

# Random Forest
rf_pred = rf.predict(X_test_scaled[:, [X.columns.get_loc(f) for f in common_features]])
rf_proba = rf.predict_proba(X_test_scaled[:, [X.columns.get_loc(f) for f in common_features]])

# XGBoost
xgb_pred = xgb.predict(X_test_scaled[:, [X.columns.get_loc(f) for f in common_features]])
xgb_proba = xgb.predict_proba(X_test_scaled[:, [X.columns.get_loc(f) for f in common_features]])

# ---- Helper functions for metrics, confusion matrix, and ROC curve ----
def print_metrics(y_true, y_pred, y_proba, model_name):
    print(f"---- {model_name} Performance ----")
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print(f"Precision: {precision_score(y_true, y_pred):.4f}")
    print(f"Recall: {recall_score(y_true, y_pred):.4f}")
    print(f"F1 Score: {f1_score(y_true, y_pred):.4f}")
    print(f"Balanced Accuracy: {balanced_accuracy_score(y_true, y_pred):.4f}")
    print(f"MCC: {matthews_corrcoef(y_true, y_pred):.4f}")
    print(f"Cohen Kappa Score: {cohen_kappa_score(y_true, y_pred):.4f}")
    print(f"Log Loss: {log_loss(y_true, y_proba):.4f}")
    print("-" * 40)

def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f"{model_name} Confusion Matrix")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()

def plot_roc_curve(y_true, y_proba, model_name):
    fpr, tpr, _ = roc_curve(y_true, y_proba[:, 1])
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f"{model_name} ROC Curve")
    plt.legend(loc="lower right")
    plt.show()

# ---- Print Metrics and Plot Confusion Matrix & ROC for each model ----
# Decision Tree
print_metrics(y_test, dt_pred, dt_proba, "Decision Tree")
plot_confusion_matrix(y_test, dt_pred, "Decision Tree")
plot_roc_curve(y_test, dt_proba, "Decision Tree")

# Random Forest
print_metrics(y_test, rf_pred, rf_proba, "Random Forest")
plot_confusion_matrix(y_test, rf_pred, "Random Forest")
plot_roc_curve(y_test, rf_proba, "Random Forest")

# XGBoost
print_metrics(y_test, xgb_pred, xgb_proba, "XGBoost")
plot_confusion_matrix(y_test, xgb_pred, "XGBoost")
plot_roc_curve(y_test, xgb_proba, "XGBoost")
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np

# Define the parameter grid for RandomizedSearchCV
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],           # Number of trees
    'max_features': ['sqrt', 'log2', None],
    'max_depth': [10, 20, 30, 40, 50, None],             # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],                     # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4],                       # Minimum number of samples required at a leaf node
    'bootstrap': [True, False],                          # Whether bootstrap samples are used when building trees
}
# Initialize the Random Forest Classifier
rf = RandomForestClassifier(random_state=42)

# Initialize StratifiedKFold for preserving class balance in each fold
strat_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize RandomizedSearchCV with StratifiedKFold
rf_random = RandomizedSearchCV(estimator=rf,
                               param_distributions=param_grid,
                               n_iter=100,    # Number of parameter settings that are sampled
                               cv=strat_kfold,  # Use StratifiedKFold cross-validation
                               verbose=2,     # Verbosity
                               random_state=42,
                               n_jobs=-1)     # Use all available cores

# Fit the model with the common selected features
rf_random.fit(X_train_scaled[:, [X.columns.get_loc(f) for f in common_features]], y_train)

# Print the best parameters
print("Best Hyperparameters found: ", rf_random.best_params_)

# Evaluate the best model on the test set using the common features
best_rf = rf_random.best_estimator_
y_pred = best_rf.predict(X_test_scaled[:, [X.columns.get_loc(f) for f in common_features]])

# Model Performance
print("Random Forest Model Performance on Test Set")
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
print(f"Classification Report:\n{classification_report(y_test, y_pred)}")

# Cross-validation scores using StratifiedKFold with the common features
cv_scores = cross_val_score(best_rf, X_train_scaled[:, [X.columns.get_loc(f) for f in common_features]], y_train, cv=strat_kfold)
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean CV Accuracy: {np.mean(cv_scores)}")
```
